```python
with open('Numbers_Dataset/Zero', 'rb') as f:
    Zero = pickle.load(f)
with open('Numbers_Dataset/One', 'rb') as f:
    One = pickle.load(f)
with open('Numbers_Dataset/Two', 'rb') as f:
    Two = pickle.load(f)
with open('Numbers_Dataset/Three', 'rb') as f:
    Three = pickle.load(f)
with open('Numbers_Dataset/Four', 'rb') as f:
    Four = pickle.load(f)
with open('Numbers_Dataset/Five', 'rb') as f:
    Five = pickle.load(f)
with open('Numbers_Dataset/Six', 'rb') as f:
    Six = pickle.load(f)
with open('Numbers_Dataset/Seven', 'rb') as f:
    Seven = pickle.load(f)
with open('Numbers_Dataset/Eight', 'rb') as f:
    Eight = pickle.load(f)
with open('Numbers_Dataset/Nine', 'rb') as f:
    Nine = pickle.load(f)
# Zero = []
# One = []
# Two = []
# Three = []
# Four = []
# Five = []
# Six = []
# Seven = []
# Eight = []
# Nine = []
```

```python
y_Zero = [0]*len(Zero)
y_One = [1]*len(One)
y_Two = [2]*len(Two)
y_Three = [3]*len(Three)
y_Four = [4]*len(Four)
y_Five = [5]*len(Five)
y_Six = [6]*len(Six)
y_Seven = [7]*len(Seven)
y_Eight = [8]*len(Eight)
y_Nine = [9]*len(Nine)
```

```python
import numpy as np

N = 19*14*4
x_Zero = np.zeros((len(Zero), N))
for i in range(len(Zero)):
#     Zero[i] = np.reshape(Zero[i],(20,14,3))
    x_Zero[i,:] = Zero[i].flatten()

x_One = np.zeros((len(One), N))    
for i in range(len(One)):
#     One[i] = np.reshape(One[i],(20,14,3))
    x_One[i,:] = One[i].flatten()

x_Two = np.zeros((len(Two),  N))    
for i in range(len(Two)):
#     Two[i] = np.reshape(Two[i],(20,14,3))
    x_Two[i,:] = Two[i].flatten()

x_Three = np.zeros((len(Three),  N))
for i in range(len(Three)):
#     Three[i] = np.reshape(Three[i],(20,14,3))
    x_Three[i,:] = Three[i].flatten()

x_Four = np.zeros((len(Four),  N))
for i in range(len(Four)):
#     Four[i] = np.reshape(Four[i],(20,14,3))
    x_Four[i,:] = Four[i].flatten()

x_Five = np.zeros((len(Five),  N))
for i in range(len(Five)):
#     Five[i] = np.reshape(Five[i],(20,14,3))
    x_Five[i,:] = Five[i].flatten()

x_Six = np.zeros((len(Six),  N))
for i in range(len(Six)):
#     Six[i] = np.reshape(Six[i],(20,14,3))
    x_Six[i,:] = Six[i].flatten()

x_Seven = np.zeros((len(Seven),  N))
for i in range(len(Seven)):
#     Seven[i] = np.reshape(Seven[i],(20,14,3))
    x_Seven[i,:] = Seven[i].flatten()

x_Eight = np.zeros((len(Eight),  N))
for i in range(len(Eight)):
#     Eight[i] = np.reshape(Eight[i],(20,14,3))
    x_Eight[i,:] = Eight[i].flatten()

x_Nine = np.zeros((len(Nine),  N))
for i in range(len(Nine)):
#     Nine[i] = np.reshape(Nine[i],(20,14,3))
    x_Nine[i,:] = Nine[i].flatten()
```

```python
from sklearn import svm
from sklearn import tree

X = np.concatenate((x_Zero, x_One, x_Two, x_Three, x_Four, x_Five, x_Six, x_Seven, x_Eight, x_Nine))
Y = y_Zero + y_One + y_Two + y_Three + y_Four + y_Five + y_Six + y_Seven + y_Eight + y_Nine

clf = svm.SVC(gamma='scale')
clf = tree.DecisionTreeClassifier()

clf.fit(X, Y)  
```

```python
correct = 0
count = 0
for i in range(len(X)):
    if clf.predict([X[i]])[0] == Y[i]:
        correct += 1
    count += 1

print(f'{correct} correct out of {count}')
```
